```python
#combining filters  
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df_filtered = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]

df_filtered
```

```python
df_filtered.shape
```

```python
# remove (USA) from event name  
```

```python
df_filtered['Event name'].str.split('(').str.get(0)
```

```python

df_filtered['Event name'] = df_filtered['Event name'].str.split('(').str.get(0)
```

```python
# Cleaning Athlete age   
```

```python
df_filtered['Athlete_age'] = 2020 - df_filtered['Athlete year of birth']
```

```python
# Removing h from time
```

```python
df_filtered['Athlete_Time'] = df_filtered['Athlete performance'].str.split('h').str.get(0)
```

```python
#changing code from manually updating to using .loc
```

```python
filtered_df = df_filtered.drop([ 'Athlete country', 'Athlete year of birth', 'Athlete age category', 'Athlete performance',],axis=1)

```

```python
# Handling NULL data
filtered_df.isnull().sum()
filtered_df= filtered_df.dropna()
# 26090-233
```

```python
#Renaming Columns without spaces
# new_cols = ['Year of event','Event dates', 'Event name', 'Event distance/length', 'Event number of finishers','Athlete gender','Athlete average speed','Athlete ID']
# filtered_df[['Year of event','Event dates', 'Event name', 'Event distance/length', 'Event number of finishers','Athlete gender','Athlete average speed','Athlete ID'] == ['Year of event','Event dates', 'Event name', 'Event distance/length', 'Event number of finishers','Athlete gender','Athlete average speed','Athlete ID'].join() ] 
```

```python
# Resetting index

filtered_df.reset_index(drop = True)
```

```python
#fix types  
```

```python
filtered_df.dtypes
```

```python
filtered_df['Athlete average speed'] = filtered_df['Athlete average speed'].astype(float)
filtered_df.dtypes
filtered_df
```

```python
#Renaming Columns   
filtered_df =  filtered_df.rename(columns = {
    'Year of event':'Year_of_event',
    'Event dates': 'Event_dates',
    'Event name': 'Event_name',
    'Event distance/length':'Event_length',
    'Event number of finishers':'Event_finishers',
    'Athlete gender':'Athlete_Gender',
    'Athlete average speed':'Athlete_avg_speed',
    'Athlete ID': 'Athlete_ID',
    'Athlete club' : 'Athlete_club'
})
```
